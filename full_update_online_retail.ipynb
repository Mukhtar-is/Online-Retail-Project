#### Detecting Outliers
An outlier is a point or set of points that are different from other points. Sometimes they can be very high or very low. It's often a good idea to detect and remove the outliers. Because outliers are one of the primary reasons for resulting in a less accurate model. Hence it's a good idea to remove them. The outlier detection and removing that I am going to perform is called IQR score technique. Often outliers can be seen with visualizations using a box plot. Shown below are the box plot of MSRP, Cylinders, Horsepower and EngineSize. Herein all the plots, you can find some points are outside the box they are none other than outliers. Let's see of there is any outliers in UnitPrice and quantity columns as they are critial variables.

```python
df.describe().T
```

```python
# finding the standard deviation of data.
df.std()
```

```python
# Let's find the quantiles of the data
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

To handle this issues, I removed the outliers using IQR method. Outliers in boxplots or using the interquartile method are determined by values that sit outside of the following criteria:
Greater than Q3 + 1.5 * IQR, or
Less than Q1 - 1.5 * IQR

In quantity column, I'll remove everything less than -13 and greater than 27. Likewise in UnitPrice, I'll remove every value less than -2 and greater than 7.

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print("------------------------------")
print(upper_bound)
```

```python
df = df[~((df["Quantity"] < lower_bound[0] )| (df["Quantity"] > upper_bound[0]))]
df = df[~((df["UnitPrice"] < lower_bound[1] )| (df["UnitPrice"] > upper_bound[1]))]
```

```python
print(df.shape)
df
```

Removed more than 62150 of outliers both quantity and UnitPrice columns. To see whether the data is now normal, let's visualize distribution of columns again. Now, it looks great with no more outliers.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['UnitPrice'])
plt.title("Distribution of UnitPrice")
plt.show()
```

##### Now, let's work time series data and Analyze the sales trends over time. 

```python
# creating datetime column
df["Invoice_Datetime"] = pd.to_datetime(df["InvoiceDate"])

# let's remove InvoiceDate column
df = df.drop('InvoiceDate', axis=1)

# create and add a year column in data
df["Year"] = df["Invoice_Datetime"].dt.year

df.head()
```

```python
# find oldest and latest invoce date
df["Invoice_Datetime"].max(), df["Invoice_Datetime"].min()
```

```python
# let's arrange columns in the data
df = df[['InvoiceNo', 'StockCode', 'Invoice_Datetime', 'Year', 'Description', 'Quantity', 'UnitPrice',
       'CustomerID', 'Country']]
df
```

5 - Identify the busiest months in terms of sales.

```python
plt.figure(figsize=(10,5))
month_sales = df.groupby(df["Invoice_Datetime"].dt.month)["Quantity"].sum()
month_sales.plot()
plt.title("Busiest months in terms of sales")
plt.ylabel("Quantity unit sales")
plt.show()
```

Sales start at around 200,000 units in January, dip slightly in February, and then gradually rise. There is a steady increase in sales from March to June, peaking slightly before dipping in July. Although there is a significant drop in sales in December, sales sharply increase from August, peaking between October and November at around 350,000 units. The peak in October-November suggests a seasonal demand, possibly due to holidays or end of year events. Some actionable insights include to focus marketing efforts in the months leading up to the peak to maximize sales, and stock up before the peak months to meet increased demand.

6- Identify the busiest days of the week in terms of sales.

```python
plt.figure(figsize=(10,5))
days_sales = df.groupby(df["Invoice_Datetime"].dt.day)["Quantity"].sum()
days_sales.plot()
plt.title("Busiest days in the month in terms of sales")
plt.xlabel("Days in month")
plt.ylabel("Quantity unit sales")
plt.show()
```

As the graph above illustrates, Sales start at around 50,000 units and show a gradual increase in the first few days. There are noticeable peaks around the 10th and 20th days of the month, where sales reach up to 110,000 units, but Sales tend to decline towards the end of the month, dropping back to around 50,000 units. Some actionable insights from this graph is the peaks around the 10th and 20th days could indicate successful promotional events or paydays when customers are more likely to make purchases, so ensuring sufficient stock is available around the mid month peaks to meet increased demand is important. Also focusing marketing efforts and Increase staffing during peak days (around the 10th and 20th days) to capitalize and handle higher sales volumes efficiently is crucial.
